# Polinomio de interpolación de lagrange

El polinomio de interpolación de lagrange es de la forma: $$\displaystyle\sum_{i=1} ^{n} \displaystyle\prod_{j \neq i} ^{n} \frac{(x-x_j)}{(x_i-x_j)} y_i$$

```python
xi=np.array([-2, 1, 4],float)
yi=np.array([3, 5, -1],float)
n = len(xi)
x = sp.symbols('x')
```

```python
if len(xi) != len(yi):
    print('los datos no son de la misma dimensión')
else:
    s = 0
    for i in range(n):
        p = 1
        for j in range(n):
            if i != j:
                p  *= (x - xi[j]) / (xi[i] - xi[j])
        s += p*yi[i]
        
```

```python
pol = sp.sympify(sp.expand(s)).evalf(4)
pol
```

```python
pol = sp.lambdify(x, s)
```

### Gráfica del polinomio con los datos

```python
x1 = np.linspace(min(xi), max(xi), 100)
plt.figure(figsize=(12,7))
plt.plot(x1, pol(x1), color='red', label=sp.sympify(p).evalf(3), lw=2.5)
plt.scatter(xi,yi, label='valores $x$', lw=3)
plt.grid()
plt.legend(loc='best', fontsize=13)
plt.title(f'Gráfica polinomio de interpolación de lagrange de grado{n-1}', fontsize=20)
```
