### Definimos a $\frac{dy}{dx} = f(x, y)$

```python
def f(x, y):
    return 3 + y/x
```

```python
h = 0.05
x0 = 1
y0 = 10

x = [x0]
y = [y0]

k = 0
while x[k] < 5:
    
#     x.append(x[k] + h)
#     y.append(y[k] + h*f(x[k], y[k]))
    
    x = np.concatenate((x, np.array([x[k] + h])), axis= -1)
    y = np.concatenate((y, np.array([y[k] + h*f(x[k], y[k])])), axis= -1)
    
    k += 1
    
plt.plot(x, y)
plt.grid()
```

### Método de Runge-Kutta de 2do orden (Regla del trapecio)

```python
h = 0.05
x0 = 1
y0 = 4

xr = [x0]
yr = [y0]

i = 0
while xr[i] < 5:
    
#     x.append(x[k] + h)
#     y.append(y[k] + h*f(x[k], y[k]))
    
    xr = np.concatenate((xr, np.array([xr[i] + h])), axis= -1)
    k1 = h*f(xr[i], yr[i])
    k2 = h*f(xr[i+1], yr[i] + k1)
    yr = np.concatenate((yr, np.array([yr[i] + 1/2*(k1+k2)])), axis= -1)
    
    i += 1
    
plt.plot(xr, yr)
plt.grid()
```
